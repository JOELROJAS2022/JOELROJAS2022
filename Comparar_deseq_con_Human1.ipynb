```r
  pequeno <- readLines("pequeno.txt", warn = FALSE)
  grande <- readLines("grande.txt", warn = FALSE)

 # Comparar si las palabras del archivo más pequeño están en el grande
  palabras_encontradas <- pequeno %in% grande

  # Crear un dataframe con los resultados
  resultado <- data.frame(
    Palabra = pequeno,
    Presente = ifelse(palabras_encontradas, "Sí", "No")
  )

  # Mostrar las palabras que Sí están en Human1
  palabras_presentes <- resultado[resultado$Presente == "Sí", ]
  palabras_presentes
```
